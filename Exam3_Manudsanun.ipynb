## วิเคราะห์ข้อมูลโดยใช้ Interquartile Range (IQR)

IGR การวิเคราะห์ข้อมูลมีความแม่นยำมากขึ้นโดยไม่ถูกบิดเบือนจากข้อมูลที่ผิดปกติ ช่วยประเมินการกระจายตัวของข้อมูลโดยไม่ต้องพึ่งพาการสมมติว่าข้อมูลต้องมีการแจกแจงปกติ ซึ่งแตกต่างจากการใช้ค่าเฉลี่ยและส่วนเบี่ยงเบนมาตรฐานที่อาจถูกบิดเบือนได้หากข้อมูลมีการกระจายตัวที่ไม่สมมาตร

### จัดการข้อมูล

```python
# เลือกเฉพาะที่มีคำว่า "มะเร็ง"
df2= df1[df1['comments'].str.contains('มะเร็ง')]
df2
```

```python
# เลือกข้อมูลที่มีจำนวนตัวอักษรมากกว่า 300 เนื่องจากเป็นการแชร์ประสบการณ์ดังนั้นข้อมูลที่ได้ควรจะมีเนื้อหาที่ยาว
df3 = df2[df2['comments'].str.len() >=300]
df3
```

```python
# เลือกเฉพาะข้อมูลที่มียอไลค์มากกว่า 0
df4 = df3[df3['ยอดไลค์'] > 0]
df4
```

```python
# เรียงข้อมูลจากน้อยไปมาก
df5 = df4.sort_values(by='ยอดไลค์')
df5
```

```python
# หาค่าไตรมาสที่ 1 (Q1)
Q1 = df4['ยอดไลค์'].quantile(0.25)
Q1
```

จะได้ว่าจำนวนยอดไลค์ที่ 25.75 คือมัธยฐานของข้อมูลส่วนแรก(25%)

```python
# หาค่าไตรมาสที่ 3 (Q3)
Q3 = df4['ยอดไลค์'].quantile(0.75)
Q3
```

จะได้ว่าจำนวนยอดไลค์ที่ 195.75 คือมัธยฐานของข้อมูลส่วนที่สาม(75%)

```python
# คำนวณ IQR ; IQR = Q3 - Q1
IQR = Q3 - Q1
IQR
```

IQR แสดงถึงการกระจายของข้อมูลกลาง 50% โดยไม่รวมค่าที่สุดโต่งหรือ outliers

```python
# การใช้ IQR ในการระบุ outliers ; [Q1 - 1.5IQR, Q3 + 1.5IQR]
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print(lower)
print(upper)
```

จะได้ว่าข้อมูลที่อยู๋นอกเหนือช่วง [-229.25 - 450.75] เป็น outliers

```python
# เลือกเาเฉพาะข้อมูลที่ไม่เป็น outliers
df5 = df4[(df4['ยอดไลค์'] >= lower) & (df4['ยอดไลค์'] <= upper)]
df5
```

```python
# เรียงลำดับข้อมูลที่ตัด outliers ออกแล้วจาก จำนวนการตอบกลับ และ ยอดไลค์
df6 = df5.sort_values(by = ['จำนวนการตอบกลับ', 'ยอดไลค์'],ascending=False)
df6.head(2)
```

```python
# ข้อมูลที่น่าสนใจข้อมูลที่ 1
print(f'ข้อมูลจากคุณ {df6.iloc[0]["name"]}'
      f'\nแสดงความคิดเห็นดังนี้ {df6.iloc[0]["comments"]}')
```

```python
# ข้อมูลที่น่าสนใจข้อมูลที่ 2
print(f'ข้อมูลจากคุณ {df6.iloc[1]["name"]}'
      f'\nแสดงความคิดเห็นดังนี้ {df6.iloc[1]["comments"]}')
```
