# 1. ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

```python
# ตัดมาเฉพาะ column ที่ต้องการ
df_1 = df[['month','station', 'Chlo_a']]
df_1.head() # เป็นการจัดการตารางเอาแค่บางคอลัมน์
```

```python
# ลบข้อมูลที่มีค่า NAN ออกจากตาราง
df_3 = df_1.dropna()
df_3
```

## ตัดเอาเฉพาะ row เพื่อแยกข้อมูลเป็น 2018 และ 2019

```python
# แยกชื่อเดือนกับปีออกจากกัน, เก็บ column "station" and "Chlo_a" ไว้ด้วย
df_split = df_3.copy()
df_split[['month', 'year']] = df_split['month'].str.split('_', expand=True, n=1) # แยกชื่อเดือนกับปีออกจากกัน
df_split
```

```python
# เลือกเฉพาะ row ที่ต้องการ ตือ ข้อมูลปี 2018
df_2018 = df_split[df_split['year']=='2018']
df_2018
```

```python
# เลือกเฉพาะ row ที่ต้องการ ตือ ข้อมูลปี 2019
df_2019 = df_split[df_split['year']=='2019']
df_2019
```

## หาค่าเฉลี่ยของละ station

```python
# ค่าเฉลี่ยของ station ในปี 2018
avg_2018 = df_2018.groupby('station')['Chlo_a'].mean()
avg_2018
```

```python
# ค่าเฉลี่ยของ station ในปี 2019
avg_2019 = df_2019.groupby('station')['Chlo_a'].mean()
avg_2019
```

```python
# ความแตกต่างค่าเฉลี่ยแต่ละ station ระหว่างปี 2018 และ 2019
diff_avg = avg_2018 - avg_2019
diff_avg
```

# 2. สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

```python
# ตัดมาเฉพาะ column ที่ต้องการ
df_2 = df[['month', 'station', 'Chlo_b']]
df_2.head()
```

```python
# ลบข้อมูลที่มีค่า NAN ออกจากตาราง
df_4 = df_2.dropna()
df_4
```

```python
# แยกชื่อเดือนกับปีออกจากกัน, เก็บ column "station" and "Chlo_b" ไว้ด้วย
df_split_2 = df_4.copy()
df_split_2[['month', 'year']] = df_split_2['month'].str.split('_', expand=True, n=1) # แยกชื่อเดือนกับปีออกจากกัน
df_split_2
```

```python
# หาค่าเฉลี่ยของ Chlo_b แต่ละ stationใช้คำสั่ง groupby() และ ใช้ unstack() เปลี่ยน index ของ dataframe ให้เป็น column
avg_chlo_b = df_4.groupby(['month', 'station'])['Chlo_b'].mean().unstack()
print("ค่าเฉลี่ยแต่ station ในแต่ละเดือนของ Chlo_b")
avg_chlo_b
```
