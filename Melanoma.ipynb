# นำเข้าข้อมูล Immuno-oncology Gene Panel

```python
Mela = pd.read_csv('/Immuno-oncology Gene Panel _Gene List.csv', encoding='latin-1')
# or 'cp1252' or any other encoding that might work for your data
Mela.head()
```

```python
Mela_HGNC = Mela.dropna(subset=['HGNC ID']) # Specify the column using the 'subset' parameter
Mela_HGNC
```

# หายีนที่ match กันในระหว่างคอลัมน์ "Gene" และ "Symbol"

```python
# แสดงเฉพาะคอลัมน์ Symbol
ncbi1 = ncbi['Symbol']

# แสดงเฉพาะคอลัมน์ Gene
Mela_HGNC1 = Mela_HGNC['Gene ']

# เปลี่ยนชื่อคอลัมน์ให้เหมือนกันเพื่อความสะดวกในการเปรียบเทียบ
Mela_HGNC1 = Mela_HGNC1.rename('Symbol')

# รวมตารางเข้าด้วยกันโดยใช้คอลัมน์ 'Symbol' เป็นตัวเชื่อม
merged_df = pd.merge(ncbi1, Mela_HGNC1, on='Symbol', how='inner')

# นับจำนวนแถวในตารางที่รวมกันแล้ว (ซึ่งก็คือจำนวนยีนที่ตรงกัน)
matched_genes_count = len(merged_df)

# แสดงผลลัพธ์
print(f"จำนวนยีนที่ตรงกันระหว่างตาราง Mela_HGNC และ ncbi คือ: {matched_genes_count} ตัว")

# แสดงตารางที่รวมกันแล้ว (ถ้าต้องการ)
merged_df
```

```python
# prompt: นำข้อมูลจาก merged_df ออกเป็นไฟล์ excel

# Specify the output file path in your Google Drive, including the file extension (.xlsx)
output_file_path = '/content/drive/MyDrive/Spatial/output_file.xlsx'  # Changed this to include .xlsx

# Export the merged DataFrame to an Excel file
merged_df.to_excel(output_file_path, index=False)  # index=False prevents writing row indices to the file

print(f"Merged data exported to: {output_file_path}")
```

จริงๆใน NCBI มีหลักการตั้งชื่อแบบ HGNC แต่ Gene name เป็นชื่อเดียวกับตั้งแบบ ENST ดังนั้นเราจะทำการ Match ข้อมูลจาก panel ทั้ง 500 gene กับ ข้อมูลจาก NCBI เลย
