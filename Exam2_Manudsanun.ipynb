## หา "person_id" ของพยาน จากข้อมูลประชากร

```python
# นำเข้าข้อมูลประชากร
person_data = '/content/drive/MyDrive/person.csv'
pr = pd.read_csv(person_data)
pr
```

  จากคำให้การณ์ของพยานพบว่า
1.   พยานคนแรกอาศัยอยู่ที่ "Northwestern Dr"
2.   พยานคนที่สองชื่อ "Annabel" อาศัยอยู่ที่ "Franklin Ave"


```python
# หา person_id ของพยานคนแรกที่อาศัยอยู่บ้านหลังสุดท้ายของถนน Northwestern Dr
wp1_id = pr[pr['address_street_name'] == 'Northwestern Dr'] # เลือกเฉพาะข้อมูลที่ถนน  Northwestern Dr
max_address = wp1_id['address_number'].idxmax()
max_address # ได้ข้อมูลเลขที่บ้าน
person_id = wp1_id.loc[max_address, 'id'] # เลือกข้อมูล person_id ที่อยู่บ้านหลังสุดท้าย
print(f'person_id ของพยานที่อาศัยอยู่ที่ Northwestern Dr คือ {person_id}')
```

```python
# หา person_id ของพยานคนชื่อ "Annabel" อาศัยอยู่ที่ "Franklin Ave"
#   แยกชื่อกับนามสกุลออกจากันเก็บ column อื่นๆไว้ด้วย
df_split = pr.copy()
df_split[['name', 'surname']] = df_split['name'].str.split(' ', expand=True, n=1)
wp2_id = df_split[(df_split['name'] == 'Annabel') & (df_split['address_street_name'] == 'Franklin Ave')] # เลือกเฉพาะข้อมูลที่ชื่อ Annabel และอาศัยอยู่ที่ Franklin Ave
person_id2 = wp2_id['id'].iloc[0] # ใช้ .iloc[0] เพื่อเลือก person_id ตัวแรกใน DataFrame
print(f'person_id ของพยานที่ชื่อ Annabel อาศัยอยู่ที่ Franklin Ave คือ {person_id2}')
```
